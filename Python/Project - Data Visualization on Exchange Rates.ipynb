# Data Visualization on Exchange Rates


## Introducing the Dataset

The dataset I use describes Euro daily exchange rates between 1999 and 2023. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the Canadian dollar is 1.5, you get 1.5 Canadian dollars if you pay 1.0 euro (one euro has more value than one Canadian dollar at this exchange rate).

Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020)  — the data source is the European Central Bank. Since the dataset gets regular updates, I downloaded it on November 2023.

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2022.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

## Data Cleaning

```python
#Rename the `[Canadian dollar ]` and `Period\Unit`: columns to something easier to type — Canadian_dollar and Time.
exchange_rates.rename(columns={'[Canadian dollar ]': 'Canadian_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)

#Change the Time column to a datetime data type.
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

#Sort the values by Time in ascending order
exchange_rates.sort_values('Time', inplace=True)

#Reset the index (and drop the initial index)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'Canadian_dollar']].copy()
euro_to_dollar['Canadian_dollar'].value_counts() # 62 '-' characters
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['Canadian_dollar'] != '-']
euro_to_dollar['Canadian_dollar'] = euro_to_dollar['Canadian_dollar'].astype(float)
euro_to_dollar.info()
euro_to_dollar.describe()
```

## Rolling Mean

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Enables Jupyter to display graphs

plt.plot(euro_to_dollar['Time'], euro_to_dollar['Canadian_dollar'])
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['Canadian_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['Canadian_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['Canadian_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
ec00_06=euro_to_dollar.copy(
                  )[(euro_to_dollar['Time'].dt.year >=2000
                    )& (euro_to_dollar['Time'].dt.year <=2006)]
ec06_15=euro_to_dollar.copy(
                  )[(euro_to_dollar['Time'].dt.year >=2006
                    )& (euro_to_dollar['Time'].dt.year <=2015)]
ec16_23=euro_to_dollar.copy(
                  )[(euro_to_dollar['Time'].dt.year >=2015)]
ec03_06=euro_to_dollar.copy(
                  )[(euro_to_dollar['Time'].dt.year >=2003
                    )& (euro_to_dollar['Time'].dt.year <=2005)]
ec06_10=euro_to_dollar.copy(
                  )[(euro_to_dollar['Time'].dt.year >=2006
                    )& (euro_to_dollar['Time'].dt.year <=2010)]
ec07_09=euro_to_dollar.copy(
                  )[(euro_to_dollar['Time'].dt.year >=2007
                    )& (euro_to_dollar['Time'].dt.year <=2009)]
ec10_15=euro_to_dollar.copy(
                  )[(euro_to_dollar['Time'].dt.year >=2010
                    )& (euro_to_dollar['Time'].dt.year <=2015)]
ec12_13=euro_to_dollar.copy(
                  )[(euro_to_dollar['Time'].dt.year >=2012
                    )& (euro_to_dollar['Time'].dt.year <=2013)]
```

```python
### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(15, 6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

### Changes to all the subplots
for ax in axes:
    ax.set_ylim(1.2, 1.7)
    ax.set_yticks([ 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5)  
    
### Ax1: 2000-2006
ax1.plot(ec00_06['Time'], ec00_06['rolling_mean'],
        color='#00868B', linewidth=2)
ax1.set_xticklabels([])
x=10800
for year in ['2000','2002','2004','2006']:
    ax1.text(x,1.17,year,alpha=0.5, fontsize=11)
    x+=730
                   
ax1.text(11800, 1.75, '2000-2006', fontsize=18, weight='bold',
        color='#00868B')


### Ax2: 2006-2015
ax2.plot(ec06_15['Time'], ec06_15['rolling_mean'],
        color='#ffa500', linewidth=2)
ax2.set_xticklabels([])
x=12950
for year in ['2006', '2008', '2010','2012', '2014','2016']:
    ax2.text(x,1.17,year,alpha=0.5, fontsize=11)
    x+=730
                   
ax2.text(14000, 1.75, '2006-2015', fontsize=18, weight='bold',
        color='#ffa500')

### Ax3: 2016-2023
ax3.plot(ec16_23['Time'], ec16_23['rolling_mean'],
        color='green', linewidth=2)
ax3.set_xticklabels([])
x=16650
for year in ['2016','2018','2020', '2022']:
    ax3.text(x,1.17,year,alpha=0.5, fontsize=11)
    x+=730
                   
ax3.text(17500, 1.75, '2016-2023', fontsize=18, weight='bold',
        color='green')


### Ax4: Entire period
ax4.plot(ec00_06['Time'], ec00_06['rolling_mean'],
        color='#00868B',linewidth=2)
ax4.plot(ec06_15['Time'], ec06_15['rolling_mean'],
        color='#ffa500',linewidth=2)
ax4.plot(ec16_23['Time'], ec16_23['rolling_mean'],
        color='green',linewidth=2)

ax4.text(14500, 1.67, '2000-2023', fontsize=18, weight='bold',
        color='#EE5C42')

### Add some transparency to the grid
ax.grid(alpha=0.5)

## Adding a title    
ax1.text(10600, 1.95, 'Euro-Canadian Dollar exchange rates between 2000 and 2023', weight='bold',
        size=20)

### Adding a signature
ax4.text(10300, 1, '©Clara Bach' + ' '*153 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```
